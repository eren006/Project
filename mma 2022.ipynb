```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import collections
import statistics
from collections import Counter
```

```python
dem = pd.read_csv("demographics.csv",index_col = 0)
dem = dem.reset_index(level=0)
dem
```

```python
tran = pd.read_csv("transactions.csv",index_col = 0)
```

```python
undis_tran = tran.loc[tran['RETAIL_DISC'] == 0]
```

```python
dis_tran = tran.loc[tran['RETAIL_DISC'] != 0]
```

```python
ut_pdid = undis_tran['PRODUCT_ID']
ut_pdsale = undis_tran[['PRODUCT_ID','SALES_VALUE']]
```

```python
# the most popular product when there is no discount
statistics.multimode(ut_pdid)
```

```python
dt_pdid = dis_tran['PRODUCT_ID']
dt_pdsale = dis_tran[['PRODUCT_ID','SALES_VALUE']]
```

```python
# the most popular product when there is a discount
statistics.multimode(dt_pdid)
```

```python
#the total sale of each product(ut)
ut_pdsale_sum = ut_pdsale.groupby('PRODUCT_ID').sum()
#the most profitable product
ut_pdsale_sum.loc[ut_pdsale_sum['SALES_VALUE'].idxmax()]
```

```python
#the total sale of each manufacturer(ut)
ut_masale = undis_tran[['MANUFACTURER','SALES_VALUE']]
ut_masale_sum = ut_masale.groupby('MANUFACTURER').sum()
#the most profitable product
ut_masale_sum.loc[ut_masale_sum['SALES_VALUE'].idxmax()]
```

```python
#the total sale of each manufacturer(dt)
dt_masale = dis_tran[['MANUFACTURER','SALES_VALUE']]
dt_masale_sum = dt_masale.groupby('MANUFACTURER').sum()
#the most profitable product
dt_masale_sum.loc[dt_masale_sum['SALES_VALUE'].idxmax()]
```

```python
undis_tran_na = tran.loc[(tran['RETAIL_DISC'] == 0)&(tran['BRAND'] != 'Private')]
```

```python
#the total sale of each manufacturer(ut_na) undiscounted + national
utna_masale = undis_tran_na[['MANUFACTURER','SALES_VALUE']]
utna_masale_sum = utna_masale.groupby('MANUFACTURER').sum()
#the most profitable product
utna_masale_sum.loc[utna_masale_sum['SALES_VALUE'].idxmax()]
utna_masale_sum
```

```python
#the total sale of each manufacturer(ut_na) undiscounted + national
utna_masale = undis_tran_na[['MANUFACTURER','COUPON_MATCH_DISC']]
utna_masale_sum = utna_masale.groupby('MANUFACTURER').sum()
#the most profitable product
utna_masale_sum.loc[utna_masale_sum['COUPON_MATCH_DISC'].idxmin()]
```

```python
#find active customer
all_c = dem
all_c = all_c.reset_index(level=0)
active = tran.groupby(tran.index).first()
active = active.reset_index(level=0)
c = all_c.merge(active, left_on='household_key', right_on='household_key', how='left')
c = c[['AGE_DESC','household_key','INCOME_DESC','HOMEOWNER_DESC','HH_COMP_DESC','HOUSEHOLD_SIZE_DESC','KID_CATEGORY_DESC','MARITAL_STATUS_CODE']]

#find percentage of each age group

#age
c['AGE_DESC'].value_counts(normalize=True) * 100
```

```python
# major age group's favorite manufacturer 
major_age = c.loc[c['AGE_DESC'] == '45-54']
major_age_buy = major_age.merge(tran, left_on='household_key', right_on='household_key', how='left')

#only national brand
major_age_buy = major_age_buy.loc[major_age_buy['BRAND'] == 'National']

#the total sale of each manufacturer
df = major_age_buy[['MANUFACTURER','SALES_VALUE']]
ds = df.groupby('MANUFACTURER').sum()
#the most profitable product
ds.loc[ds['SALES_VALUE'].idxmax()]
```

```python
#major age group's favorite product type
df = major_age_buy[['COMMODITY_DESC','SALES_VALUE']]
ds = df.groupby('COMMODITY_DESC').sum()
#the most profitable product type
ds.sort_values(by='SALES_VALUE', ascending=False)
```

```python
#the best soft drinks manufacturer 
df = tran[['MANUFACTURER','COMMODITY_DESC','SALES_VALUE','BRAND']]
df = df.loc[(df['COMMODITY_DESC'] == 'SOFT DRINKS') & (df['BRAND'] == 'National')]
ds = df.groupby('MANUFACTURER').sum()
#the most profitable product
ds.loc[ds['SALES_VALUE'].idxmax()]
```
